The most considering issue is that when the synthetic data is generating form ture samples there might be a issue of information loss if the synthetic samples doesn't represents the actual samples properly.



# Undersampling 

By removing datapoint from the majority class may often faced an issue of loss of valuable information because a huge number of datapoints as removing from the dataset. As the data points are sampling respect to the minority class so the model will work more sensitively for the minority class this will create an issue with the majority class where model will perform with it poorly thus the model will be baised. # Cross_validation



Cross-validation is a type resampling procedure used to evaluate machine learning models on a limited data sample. It is a technique used to protect against overfitting in a predictive model, particularly in a case where the amount of data may be limited.

# Types of Cross Validation


# K-Fold Cross Validation

In K_Fold Cross Validation the performance and generalization ability of a model is assessed. K_Fold Cross Validation technique is more useful when we are dealing with limited datapoints, where the data original dataset is divided into K number of subsets or folds and each folds containing the same amount of data points as well.Each fold is used as a testing set exactly once, while the remaining k-1 folds are used as training data.For the each unique group of k-fold a group is taken as hold out or test set where the remaining groups are considered as training set. Then it fit the model on the training set and evaluate it on the test set and retain the evaluation score and discard
the model.

However, the common choices for k in k-fold cross-validation are 5, 10, or even higher values, depending on the size of the dataset and computational resources available. However, smaller values of k may result in higher variance in the performance estimate, while larger values may increase the computational cost.

# Stratified Cross Validation

Stratified k-fold cross-validation is geneal variation for the  k-fold cross-validation aims to focused specially these class having imbalance class in a dataset. In Stratified k-fold technique the class distribution will be same as like as the original dataset. In this technique the datasets are divided as just like the traditional k-fold cross-validation. In Stratified Cross Validation the data distribution is preserved that means if the provided dataset is imbalanced then all the fold will also have the similar amount of imbalance ratio as like the original dataset. The training as testing stages are performing as like the traditional k-fold cross-validation technique.



# Leave-One-Out Cross-Validation (LOOCV)


Leave-One-Out Cross-Validation (LOOCV) is another type of a special case of k-fold cross-validation where k is equal to the number of datapoint presents into the dataset.  In this approach a model is trained over the all the other datapoints expect one. That datapoints is used a test point. After testing over the single datapoint the performance are then avaraged across all datapoints for obtaining a single test result for the proposed model.




There is some advantages to work the Leave-One-Out Cross-Validation (LOOCV) these are 
**High Utilization of Data and Low Bias**
beacuse in  Leave-One-Out Cross-Validation we makes the maximum use of the available data we have for both training and testing peocess, since we olny use one datapoint to test and use all the other datapoints in training phase.

Although it has some advantages but there is also be some disadvantages regarding uses of the  Leave-One-Out Cross-Validation and these are 
**High Variance and Computational Cost** 
Leave-One-Out Cross-Validation is introduced us with a large Computational Cost when we are dealing with a large amonut of data, since the training and testing process is repeated for all the single datapoints seperately. 

However this apporoach can be useful for us if we are dealing with a moderate and small dataset and wish to have an unbaised model performance.
